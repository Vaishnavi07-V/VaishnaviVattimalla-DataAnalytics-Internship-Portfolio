```python
import pandas as pd

# Load dataset
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset (3).xlsx")

# Display first 5 rows
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
df.isnull().sum()
```

```python
# Fill Age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill City with mode
df["City"] = df["City"].fillna(df["City"].mode()[0])

# Check again
print(df.isnull().sum())
```

```python
df.info()
```

###Statistical Validation

```python
from scipy.stats import ttest_ind

male = df[df["Gender"]=="Male"]["Total_Sales"]
female = df[df["Gender"]=="Female"]["Total_Sales"]

t,p = ttest_ind(male,female)

print("T-statistic:",t)
print("P-value:",p)
```

######If P-value = 0.02
######Since the p-value is less than 0.05, we reject the null hypothesis. There is a statistically significant difference in spending between male and female customers.
######If P-value = 0.32
######Since the p-value is greater than 0.05, we fail to reject the null hypothesis. There is no statistically significant difference in spending between male and female customers.

### Statistical Validation

#####To validate the findings statistically, an Independent Samples T-Test was performed.

#####Null Hypothesis (H₀): There is no significant difference in the average purchase amount between male and female customers.

#####The p-value obtained from the T-Test was compared with the significance level of 0.05.

#####If p-value < 0.05: Reject the null hypothesis.

#####If p-value ≥ 0.05: Fail to reject the null hypothesis.
